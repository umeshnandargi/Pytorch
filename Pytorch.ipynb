##Slices

```python
x = my_tesnor[1:].reshape(3,3)
print(x)
x_col_slice = x[:,1]
print(x_col_slice)
x_col = x_col_slice.reshape(3,1)
print(x_col)
```

## Tensor Math operations

```python
a = torch.tensor([1,2,3,4,5])
b = torch.arange(6,11)
a,b
```

```python
#Addition
a + b
torch.add(a,b)
```

```python
# Subtraction
a-b
torch.sub(a,b)
```

```python
# Multiplication

a*b
torch.mul(a,b)
```

```python
# Division
a/b
torch.div(a,b)
```

```python
# Remainder modulus

b % a
torch.remainder(b, a)
```

```python
# exponent
b ** a
torch.pow(b,a )
```

```python
# Reassignment
a = a+b
a.add_(b)
```
